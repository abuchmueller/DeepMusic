### Generate the Access Token for Spotify

```python
# this function generates an access token, your client_id and your client_secret have to be inserted into the function, those are 
# gotten from the spotify-dashboard

def generate_token():
    credentials = oauth2.SpotifyClientCredentials(
        client_id="YOUR CLIENT ID",
        client_secret="YOUR CLIENT SECRET")
    token = credentials.get_access_token()
    return token


token = generate_token()
sp = spotipy.Spotify(auth=token)
```
